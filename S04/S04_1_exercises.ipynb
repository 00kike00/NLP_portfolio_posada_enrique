```python
# Corpus
s1 = "<s> I am Sam </s>"
s2 = "<s> Sam I am </s>"
s3 = "<s> I do not like green eggs and ham </s>"

corpus = [s1, s2, s3]
```

```python
probabilities = {}
words_frequency = {}
bigrams_frequency = {}


for s in corpus:
    i = 0
    words = s.split()
    while i < len(words):
        if i != 0:
            bigram = words[i-1] + " " + words[i]
            if bigrams_frequency.get(bigram) == None:
                bigrams_frequency[bigram] = 1
            else: 
                bigrams_frequency[bigram] = bigrams_frequency.get(bigram) + 1
        if words_frequency.get(words[i]) == None:
            words_frequency[words[i]] = 1
        else: 
            words_frequency[words[i]] = words_frequency.get(words[i]) + 1
        i+=1
for bgram, freq in bigrams_frequency.items():
    first_word = bgram.split()[0]
    probabilities[bgram] = freq / words_frequency[first_word]

print(probabilities)
```
