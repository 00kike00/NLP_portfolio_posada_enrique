```python
# 4
def rev(s: str) -> str:
    s_inv = ""
    for i in range(len(s), 0, -1):
        s_inv += s[i-1]
    
    return s_inv
string = "hello"
print(rev(string))
```

```python
# 6
def anagram(s1: str, s2: str):
    if len(s1) != len(s2):
        return False
    else:
        cs = {}
        for c1, c2 in zip(s1, s2):
            if cs.get(c1) == None:
                cs[c1] = 1
            else: 
                cs[c1] = cs.get(c1) + 1
            if cs.get(c2) == None:
                cs[c2] = -1
            else: 
                cs[c2] = cs.get(c2) - 1
        for _, v in cs.items():
            if v != 0:
                return False

    return True

string1 = "canvaci"
string2 = "narciso"

print(anagram(string1, string2))
```

```python
# 7
def del_vowels(string: str) -> str:
    new_string = ""
    for char in string:
        if char.lower() not in ('a', 'e', 'i', 'o', 'u'):
            new_string+=char
        
    return new_string    
string = "murcielago"

strin = del_vowels(string)
print(strin)
```

```python
# 8
def palindrome(s: str) -> bool:
    if s == rev(s):
        return True
    return False 

string = "detartrated"
print(palindrome(string))
```
